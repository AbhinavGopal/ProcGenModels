```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
```

# Plotting Mean Reward for 50 Training Levels

```python
#pass in appropriate paths to baseline, reward shaping, larger convolution, and reward + convolution csvs.
df = pd.read_csv('/content/drive/MyDrive/CS 182/Trained Models/baseline-50/progress.csv')
df_rew = pd.read_csv('/content/drive/MyDrive/CS 182/Trained Models/baseline-50-reward/progress.csv')# Reward Shaping csv
df_conv = pd.read_csv('/content/drive/MyDrive/CS 182/Trained Models/baseline-50-conv/progress.csv') # Large Conv csv
df_rewconv = pd.read_csv('/content/drive/MyDrive/CS 182/Trained Models/baseline-50-conv-reward/progress.csv') # Reward Shaping + Large Conv csv
df_reward_rms = pd.read_csv('/content/drive/MyDrive/CS 182/Trained Models/baseline-50-reward-rms/progress.csv') # Reward Shaping + Large Conv csv

total_timesteps = df['misc/total_timesteps']

#case for if we want log_timesteps
#get the lowest timesteps possible
if len(df_rew['misc/total_timesteps']) < len(total_timesteps):
  total_timesteps = df_rew['misc/total_timesteps']
if len(df_conv['misc/total_timesteps']) < len(total_timesteps):
  total_timesteps = df_conv['misc/total_timesteps']
if len(df_rewconv['misc/total_timesteps']) < len(total_timesteps):
  total_timesteps = df_rewconv['misc/total_timesteps']
if len(df_reward_rms['misc/total_timesteps']) < len(total_timesteps):
  total_timesteps = df_reward_rms['misc/total_timesteps']
```

```python
total_log_timesteps = moving_average(total_timesteps.to_numpy(), 20)
eval_baseline_reward = moving_average(df['eval_eprewmean'][0:len(total_timesteps)], 20) 
eval_rew_reward = moving_average(df_rew['eval_eprewmean'][0:len(total_timesteps)], 20) 
eval_conv_reward = moving_average(df_conv['eval_eprewmean'][0:len(total_timesteps)], 20) 
eval_rewconv_reward = moving_average(df_rewconv['eval_eprewmean'][0:len(total_timesteps)], 20)
eval_rms_reward = moving_average(df_reward_rms['eval_eprewmean'][0:len(total_timesteps)], 20)
```

```python
plt.figure(figsize=(9,6))
plt.plot(total_log_timesteps, eval_baseline_reward, label = 'baseline')
plt.plot(total_log_timesteps, eval_rew_reward, label = 'reward shaping')
plt.plot(total_log_timesteps, eval_conv_reward, label = 'larger convolution')
plt.plot(total_log_timesteps, eval_rewconv_reward, label = 'reward shaping + larger convolution')
plt.plot(total_log_timesteps, eval_rms_reward, label = 'reward shaping + RMSProp')

plt.title('Evaluation rewards in FruitBot for 50 levels of training')
plt.xlabel('Timesteps')
plt.ylabel('Mean Reward')
plt.legend(loc='lower right')
plt.savefig("50-levels.png");
```
